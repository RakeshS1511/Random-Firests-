## COMPANY DATA

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# loading data 
data = pd.read_csv("Company_Data.csv")
data.head()
```

```python
data.loc[data.Sales >=8.5,'sales']= 'high'
data.loc[data.Sales <8.5,'sales']= 'low'
```

```python
data.drop('Sales',1,inplace=True)
```

```python
data.isnull().sum()
```

```python
from sklearn import preprocessing 
label_encoder = preprocessing.LabelEncoder() 
data['US']= label_encoder.fit_transform(data['US']) 
data['US'].unique() 
```

```python
label_encoder = preprocessing.LabelEncoder() 
data['Urban']= label_encoder.fit_transform(data['Urban']) 
```

```python
label_encoder = preprocessing.LabelEncoder() 
data['ShelveLoc']= label_encoder.fit_transform(data['ShelveLoc'])
```

```python
colnames = list(data.columns)
predictors = colnames[0:10]
target = colnames[10]
```

```python
predictors
```

```python
target
```

```python
label_encoder = preprocessing.LabelEncoder() 
data['sales']= label_encoder.fit_transform(data['sales']) 
```

```python
X = data[predictors]
Y = data[target]
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_jobs=2,oob_score=True,n_estimators=1000,criterion="entropy")
```

```python
rf.fit(X,Y)
```

```python
rf.estimators_
```

```python
rf.classes_
```

```python
rf.n_classes_
```

```python
rf.n_classes_
```

```python
rf.n_outputs_
```

```python
data['rf_pred'] = rf.predict(X)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(data['sales'],data['rf_pred'])
```

### Now will try with 500 decision tress

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_jobs=2,oob_score=True,n_estimators=500,criterion="entropy")
```

```python
rf.fit(X,Y)
```

```python
data['rf_pred2'] = rf.predict(X)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(data['sales'],data['rf_pred2'])
```

## FRAUD CHECK

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# loading data 
data = pd.read_csv("Fraud_check.csv")
data.head()
```

```python
data.loc[data['Taxable.Income'] <=30000,'Taxable_Income']= 'Risky'
data.loc[data['Taxable.Income'] >30000,'Taxable_Income']= 'Good'
```

```python
data.head(2)
```

```python
data.drop('Taxable.Income',1,inplace=True)
```

```python
data['Marital.Status'].value_counts()
```

```python
from sklearn import preprocessing 
label_encoder = preprocessing.LabelEncoder() 
data['Marital.Status']= label_encoder.fit_transform(data['Marital.Status']) 
data['Marital.Status'].unique() 
```

```python
data.head(2)
```

```python
data.Undergrad.value_counts()
```

```python
data['Undergrad']= label_encoder.fit_transform(data['Undergrad']) 
data['Undergrad'].unique() 
```

```python
data.Urban.value_counts()
```

```python
data['Urban']= label_encoder.fit_transform(data['Urban']) 
data['Urban'].unique() 
```

```python
data.head()
```

```python
colnames = list(data.columns)
predictors = colnames[0:5]
target = colnames[5]
```

```python
predictors
```

```python
target
```

```python
X = data[predictors]
Y = data[target]
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_jobs=2,oob_score=True,n_estimators=1000,criterion="entropy")
rf.fit(X,Y)
```

```python
data.head()
```

```python
data['rf_pred'] = rf.predict(X)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(data['Taxable_Income'],data['rf_pred'])
```
